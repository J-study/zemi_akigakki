```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image

device = "cuda" if torch.cuda.is_available() else 'cpu'

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_device=torch.float16
).to(device)

#処理軽減用
pipe.enable_attention_slicing()
pipe.enable_vae_tiling()

pipe.load_lora_weights(
    "/content/drive/MyDrive/Loras/zemi_notXL/output",
    weight_name = "zemi_notXL-10.safetensors"
)

item_image_path = "/content/drive/MyDrive/white_margin_change_2.png"
init_image = Image.open(item_image_path).convert("RGB").resize((512,512))

#prompt = "full-body shot, a high-quality photo of a man wearing clothes, masterpiece, best quality"
prompt = "a full-body photo of a stylish man wearing the uploaded clothing item, casual fashion, realistic lighting, high quality, neutral background"
negative_prompt = "anime, illustration,low quality, cutoff, cropped, nsfw, blurry,  bad hands, bad anatomy"

image = pipe(
    prompt = prompt,
    image = init_image,
    strength = 0.9,
    guidance_scale = 9,
    num_inference_steps = 30,
    negative_prompt = negative_prompt,
).images[0]

output_path  = "/content/drive/MyDrive/output_zemi_ver4.2.png"
image.save(output_path)

```
